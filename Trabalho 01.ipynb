## Nesse notebook, vamos analisar a quantidade de sorteios por ano, a quantidade de ganhadores por ano e o percentual de ganhadores por sorteios, apenas utilizando os dados da sena.

```python
#Nesse bloco, os dados são lidos e atribuidos a "Loteria"
import pandas as pd
Loteria = pd.read_csv("basedados.csv")
Loteria = Loteria.drop("Unnamed: 0", axis = 1)
```

```python
#Nesse bloco, crio um dicionário que contém a quantidade de sorteios em cada ano
Data = Loteria["Data Sorteio"]
anos = []
years = {}
for i, data in enumerate(Data):
    partes = Data[i].split()
    ano = partes[0].split("-")
    anos.append(ano[0])
for i in anos:
    years[i] = 0
for i in anos:
    if i in years:
        years[i] += 1
print(years)
```

```python
#Nesse bloco, crio um dicionário que contém a quantidade de ganhadores por ano
Sena = Loteria["Ganhadores_Sena"]
sena = {}
for i in anos:
    sena[i] = 0
for i, value in enumerate(anos):
    if value in sena:
        sena[value] += Sena[i]
print(sena)
```

```python
#Nesse bloco, crio um gráfico para mostrar a quantidade de sorteios por ano usando a biblioteca matplotlib.pyplot
year = []
sorteios = []
for key, value in years.items():
    year.append(key)
    sorteios.append(value)
    
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(year, sorteios)
plt.xticks(rotation = 90)
plt.xlabel("Ano")
plt.ylabel("Sorteios")
plt.title("Quantidade de sorteios por ano")
plt.show()
```

```python
#Nesse bloco, crio um gráfico para mostrar a quantidade de ganhadores por ano usando a biblioteca matplotlib.pyplot
year = []
ganhadores = []
for key, value in sena.items():
    year.append(key)
    ganhadores.append(value)
    
plt.plot(year, ganhadores)
plt.xticks(rotation = 90)
plt.xlabel("Ano")
plt.ylabel("ganhadores")
plt.title("Quantidade de Ganhadores por ano")
plt.show()
```

```python
#Nesse bloco, mostro o percentual de ganhadores pela quantidade de sorteios em cada ano
percentual = {}
percento = []
for i, value in enumerate(ganhadores):
    percento.append(ganhadores[i]*100/sorteios[i])
for i in anos:
    percentual[i] = 0
for i, value in enumerate(sena):
    if value in percentual:
        percentual[value] = percento[i]
print(percentual)
```
